```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lowpass_df = pd.read_csv("/Users/anhducduong/Documents/GitHub/MainPipelineRepo/combined_data_timewindow_2min_labels_valence_arousal_lowpass.csv").copy()
no_lowpass_df = pd.read_csv("/Users/anhducduong/Documents/GitHub/MainPipelineRepo/combined_data_timewindow_2min_labels_valence_arousal.csv").copy()
```

```python
# Filter the data for the first group of the first participant
group_id = 1
participant_id = 1
lowpass_x = lowpass_df[(lowpass_df['participantId'] == participant_id) & (lowpass_df['groupid'] == group_id)]['x']
no_lowpass_x = no_lowpass_df[(no_lowpass_df['participantId'] == participant_id) & (no_lowpass_df['groupid'] == group_id)]['x']

# Plot the data
plt.figure(figsize=(14, 7))
plt.plot(lowpass_x.values, label='Lowpass X Data')
plt.plot(no_lowpass_x.values, label='No Lowpass X Data')
plt.xlabel('Time')
plt.ylabel('X Data')
plt.title('Comparison of X Data for Group 1 of Participant 1')
plt.legend()
plt.show()
```

```python
# Filter the data for the first 500 samples of the first group of the first participant
group_id = 1
participant_id = 1
lowpass_x_500 = lowpass_df[(lowpass_df['participantId'] == participant_id) & (lowpass_df['groupid'] == group_id)]['x'].head(200)
no_lowpass_x_500 = no_lowpass_df[(no_lowpass_df['participantId'] == participant_id) & (no_lowpass_df['groupid'] == group_id)]['x'].head(200)

# Plot the data
plt.figure(figsize=(14, 7))
plt.plot(lowpass_x_500.values, label='Lowpass X Data (First 500 Samples)')
plt.plot(no_lowpass_x_500.values, label='No Lowpass X Data (First 500 Samples)')
plt.xlabel('Time')
plt.ylabel('X Data')
plt.title('Comparison of X Data for First 500 Samples of Group 1 of Participant 1')
plt.legend()
plt.show()
```
