```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class AggregateTimeBins(BaseEstimator, TransformerMixin):
    def __init__(self, bin_length='5min'):
        self.bin_length = bin_length

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # Resample the data based on the specified bin length
        X['datetime'] = pd.to_datetime(X['datetime'], unit='ms')
        aggregated_data = X.set_index('datetime').resample(self.bin_length).mean()
        
        # Add back reportId, context, arousal, valence for each time bin
        aggregated_data['reportId'] = X['reportId'].iloc[0]  # Assuming the reportId is the same within the bin
        aggregated_data['context'] = X['context'].iloc[0]    # Assuming the context is the same within the bin
        aggregated_data.reset_index(inplace=True)

        print(f"Data aggregated successfully into {self.bin_length} bins.")
        return aggregated_data
```
