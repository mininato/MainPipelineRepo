LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder

# Convert the target variable from string to integer using label encoding
label_encoder = LabelEncoder()
arousal_valence_encoded = label_encoder.fit_transform(arousal_valence)
arousal_valence_encoded = pd.DataFrame(arousal_valence_encoded, columns=['Encoded'])

print("Arousal Valence Encoded:", arousal_valence_encoded)

# Get the unique classes and their corresponding encoded values
classes = label_encoder.classes_
encoded_values = label_encoder.transform(classes)

# Create a dictionary to map encoded values to original strings
encoding_map = dict(zip(encoded_values, classes))

print("Encoding map:", encoding_map)
```

```python
import time
from sklearn.model_selection import StratifiedGroupKFold, cross_val_predict
from xgboost import XGBClassifier
from skopt import BayesSearchCV
from sklearn.metrics import classification_report, accuracy_score, f1_score, precision_score, recall_score

# Define the number of folds
n_splits = 5

# Create the StratifiedGroupKFold object
sgkf = StratifiedGroupKFold(n_splits=n_splits)

# Create XGBClassifier instance
xgb = XGBClassifier(objective='multi:softmax', num_class=9, random_state=42)

fold = 1

# Define the parameter search space
param_space = {
    'learning_rate': (0.01, 0.3, 'log-uniform'),
    'n_estimators': (100, 1000),
    'max_depth': (3, 10),
    'min_child_weight': (1, 10),
    'subsample': (0.5, 1.0),
    'colsample_bytree': (0.5, 1.0),
    'gamma': (0, 10),
    'reg_alpha': (0, 10),  # alpha in xgboost is reg_alpha
    'reg_lambda': (0, 10), # lambda in xgboost is reg_lambda
}

# Initialize BayesSearchCV
opt = BayesSearchCV(
                estimator=xgb,
                search_spaces=param_space, 
                cv=sgkf, 
                random_state=42, 
                verbose=2, 
                n_jobs=-1,
                n_iter=30, 
                n_points=1,
                pre_dispatch='2*n_jobs',
                scoring='accuracy'
                )  
print("BayesSearchCV initialized")

# Start the timer
start_time = time.time()
print("timer started")

# Perform hyperparameter optimization
opt.fit(scaled_data, arousal_valence_encoded, groups=groups)

# Stop the timer
end_time = time.time()

print("end timer")
print(f"optimizer fit")

# Calculate the total time taken
total_time = end_time - start_time

# Access the cross-validation results
cv_results = opt.cv_results_

# Get the best estimator
best_est = opt.best_estimator_

print(f"Best parameters found: {opt.best_params_}")
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score, precision_score, recall_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

# Get cross-validated predictions on the train set for evaluation
y_pred = cross_val_predict(best_est, scaled_data, arousal_valence_encoded, groups=groups, cv=sgkf)

accuracy = accuracy_score(arousal_valence_encoded, y_pred)
f1 = f1_score(arousal_valence_encoded, y_pred, average='macro')
precision = precision_score(arousal_valence_encoded, y_pred, average='macro')
recall = recall_score(arousal_valence_encoded, y_pred, average='macro')


# Compute the confusion matrix
cm = confusion_matrix(arousal_valence_encoded, y_pred)

# Print the confusion matrix
print("Confusion Matrix:")
print(cm)
```
